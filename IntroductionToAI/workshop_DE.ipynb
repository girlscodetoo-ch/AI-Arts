````python
def read_image(filepath):
    image = tf.io.read_file(filepath) # read image
    image = tf.image.decode_image(image, channels=3) # decode image
    image = tf.image.convert_image_dtype(image, tf.float32) # rescale the pixels
    return image
````

````python
def rescale_image(image, max_dim=512):
    shape = tf.cast(tf.shape(image)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim
    new_shape = tf.cast(shape * scale, tf.int32)
    image = tf.image.resize(image, new_shape)
    image = image[tf.newaxis, :]
    # TODO: alternative line here just in case we want to make all figures square
    # image = tf.image.resize_with_crop_or_pad(image, max_dim, max_dim)
    print("Rescaling image from %.0f x %.0f to shape %s x %s"%(*tuple(shape.numpy()), *tuple(new_shape.numpy())))
    return image
````

````python
# read photo and art
````

````python
content = read_image('../images/dog.jpg')
````

````python
content = rescale_image(content)
````

````python
style = read_image('../images/kandinsky.jpg')
````

````python
style = rescale_image(style)
````

````python
# use tensorflow hub
````

````python
import tensorflow_hub as tf_hub

hub_model = tf_hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')

stylized_image = hub_model(tf.constant(content), tf.constant(style))[0]
````
